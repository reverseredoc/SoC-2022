```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random as rnd
from sklearn import datasets,linear_model
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
```

```python
data1=pd.read_csv(r"/home/user/Downloads/weatherHistory.csv")#r is for raw string 
df=pd.DataFrame(data1)
df.head(n=6) # 6 rows printed out
```

```python
x=df.drop('Humidity',1)#dropped from x axis
y=df['Humidity']# Humidity is y our dependent variable 
df=df.drop(['Formatted Date','Summary','Precip Type','Loud Cover','Daily Summary','Apparent Temperature (C)','Visibility (km)','Wind Bearing (degrees)','Wind Speed (km/h)','Humidity'],axis=1)
```

```python
df.head(n=5)
```

```python
#create test and train variables using test_train_split from sklearn library
x_train,x_test,y_train,y_test=train_test_split(df,y,test_size=0.25)
lrm=linear_model.LinearRegression()
lrm_model=lrm.fit(x_train,y_train)
lrm_predictions=lrm.predict(x_test)
#plotting the prdictions with a scatter plot
plt.scatter(y_test,lrm_predictions)
plt.title("Plot of predicted versus Actual Values")
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
```

```python
print('Score:',lrm_model.score(x_test,y_test))
```
